custom `FeatureExtractor` module.

### Steps Performed

- Loaded the cleaned phishing URL dataset
- Imported the reusable feature extraction logic from `src/feature_extraction.py`
- Iterated through each URL in the dataset
- Extracted a fixed set of numerical features for every URL
- Constructed the feature matrix `X` and label vector `y`


```python
df = pd.read_csv("../datasets/phishing_url_cleaned.csv")
```

```python
x = []
y = []
for url,label in zip(df['URL'],df['label']):
    obj = FeatureExtractor(url)
    features = obj.feature_vector()
    x.append(features)
    y.append(label)
```

```python
import numpy as np
x = np.array(x)
y = np.array(y)
```
